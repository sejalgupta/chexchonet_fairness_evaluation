AUC over 5 runs

```python
Eval1 = pd.read_pickle("./results/results9/ret_mets_test_results_9.csv_updated.pkl")
Eval2 = pd.read_pickle("./results/results21/ret_mets_test_results_21.csv_updated.pkl")
Eval3 = pd.read_pickle("./results/results25/ret_mets_test_results_25.csv_updated.pkl")
Eval4 = pd.read_pickle("./results/results35/ret_mets_test_results_35.csv_updated.pkl")
Eval5 = pd.read_pickle("./results/results42/ret_mets_test_results_42.csv_updated.pkl")
```

```python
evals = [Eval1, Eval2, Eval3, Eval4, Eval5]

# Initialize DataFrame
# Assuming each eval has the same set of metrics
metrics = list(evals[0]['ret_met'].keys()) + ['no finding']  # Adding 'No Finding' as a separate metric
columns = [f'auc{i+1}' for i in range(len(evals))]  # Generate column names
Evalall = pd.DataFrame(index=metrics, columns=columns)  # Use metrics as index
EvalallAuprc = pd.DataFrame(index=metrics, columns=[f'auprc{i+1}' for i in range(len(evals))])  # Use metrics as index

# Populate AUC values
for j, eval_data in enumerate(evals, start=1):
    for metric, values in eval_data['ret_met'].items():
        auc_value = values['auroc'][0]  # Using 'auroc' tuple's first element
        auprc_value = values['auprc'][0]  # Using 'auroc' tuple's first element
        Evalall.at[metric, f'auc{j}'] = auc_value
        EvalallAuprc.at[metric, f'auprc{j}'] = auprc_value
        if metric == 'composite':
            # Calculate and store "No Finding" AUC as 1 minus the "Composite" AUC
            Evalall.at['no finding', f'auc{j}'] = 1 - auc_value
            EvalallAuprc.at['no finding', f'auprc{j}'] = 1 - auprc_value

# Convert all AUC columns to numeric and round
Evalall.iloc[:, 1:] = Evalall.iloc[:, 1:].apply(pd.to_numeric).round(3)
EvalallAuprc.iloc[:, 1:] = EvalallAuprc.iloc[:, 1:].apply(pd.to_numeric).round(3)
```

```python
def log_stats(eval_1, eval_name):
    print(f'ALL Mean of 4 {eval_name} mean over 5 run: {round(eval_1.mean(axis=0).mean(),3)}')
    print(f'ALL Confidence interval of 4 {eval_name} mean over 5 run: {round(1.96 * eval_1.mean(axis=0).std()/np.sqrt(5) ,3 )}')
    print(f'ALL Mean of {eval_name} per disease over 5 run:\n{round(eval_1.mean(axis=1),3)}')
    print(f"ALL Confidence interval of {eval_name} per disease over 5 run:\n{round(1.96 * eval_1.std(axis=1)/ np.sqrt(5),3)}")
```

```python
log_stats(Evalall, "AUC")
```

```python
log_stats(EvalallAuprc, "AUPRC")
```
